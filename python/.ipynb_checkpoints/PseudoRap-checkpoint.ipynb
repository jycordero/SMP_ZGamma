```python
import numpy as np
import matplotlib.pylab as plt
```

```python
theta = np.arange(0.01,np.pi/2, 0.01)
eta = -np.log(np.tan(theta/2))
Deta = [[0 for _ in eta] for _ in eta]
for i in range(len(theta)):
    for j in range(len(theta)):
        Deta[i][j] = -np.log(np.tan(theta[i]/2))+np.log(np.tan(theta[j]/2))
        if np.abs(Deta[i][j]) > 2.4:
            Deta[i][j] = 0
plt.figure()
plt.plot(theta*(180/np.pi),eta)


#X,Y = np.meshgrid(eta,eta)
X,Y = np.meshgrid(theta*(180/np.pi),theta*(180/np.pi))
plt.figure(figsize = (10,10))
plt.contourf(X,Y,Deta,30,cmap='RdGy')
plt.colorbar()
plt.show()
```

```python
def dR(theta,theta1,phi,phi1):
    eta = ETA(theta)
    eta1 = ETA(theta1)
    return(np.sqrt((eta-eta1)**2 + (phi-phi1)**2))
```

```python
def ETA(theta):
    return -np.log(np.tan(theta/2))
```

```python
theta = np.arange(0.01,np.pi/2, 0.01)

D_phi = np.arange(0.01,np.pi, 0.01)
D_eta = np.arange(0.01,5.5, 0.01)
X,Y = np.meshgrid(D_phi,D_eta)
D_R = [[0 for _ in D_phi] for _ in D_eta]
for i in range(len(D_eta)):
    for j in range(len(D_phi)):
        D_R[i][j] = np.sqrt(D_eta[i]**2 + D_phi[j]**2)

plt.figure(figsize=(10,10))
ax = plt.gca()
ax.set_xlabel(r"$\Delta\phi$")
ax.set_ylabel(r"$\Delta\eta$")
plt.contourf(X,Y,D_R,30,cmap = "RdGy")
plt.colorbar()

#------------------------------------------------
#------------------------------------------------

phi = np.arange(0.01,np.pi, 0.01)
theta0 = 90
theta = np.arange(0.01,180, 1)
#print(theta-theta0*(np.pi/180))

X,Y = np.meshgrid(phi*(180/np.pi),theta)
D_R = [[0 for _ in phi] for _ in theta]
for i in range(len(theta)):
    for j in range(len(D_phi)):
        #D_R[i][j] = np.sqrt(D_eta[i]**2 + D_phi[j]**2)
        D_R[i][j] = dR((theta0+0.01)*(np.pi/180.0),
                       theta[i]*(np.pi/180.0),
                       0,
                       phi[j])
        if(D_R[i][j] < 3.1 and  D_R[i][j] > 2.9): D_R[i][j] = 100
        elif(D_R[i][j] < 0.2 ): D_R[i][j] = -100
        #elif(D_R[i][j] > 5.6 ): D_R[i][j] = 0
        elif(np.abs(ETA(theta[i]*(np.pi/180.0))) < 2.4 ): D_R[i][j] = 0
        else: D_R[i][j] = 50

plt.figure(figsize=(10,10))
ax = plt.gca()
ax.set_title("$\Delta$ R")
ax.set_xlabel(r"$\phi(\phi_0$ = 0)")
ax.set_ylabel(r"$\theta(\theta_0$ = 90)")
plt.contourf(X,Y,D_R,60)#,cmap = "RdGy")
plt.colorbar()
plt.show()
```

```python
#------------------------------------------------
#------------------------------------------------

phi = np.arange(0.01,np.pi, 0.01)
theta0 = 90
theta = np.arange(0.01,180, 1)
#print(theta-theta0*(np.pi/180))

X,Y = np.meshgrid(phi*(180/np.pi),theta)
D_R = [[0 for _ in phi] for _ in theta]
for i in range(len(theta)):
    for j in range(len(D_phi)):
        #D_R[i][j] = np.sqrt(D_eta[i]**2 + D_phi[j]**2)
        D_R[i][j] = dR((theta0+0.01)*(np.pi/180.0),
                       theta[i]*(np.pi/180.0),
                       0,
                       phi[j])
        #if(D_R[i][j] < 3.1 and  D_R[i][j] > 2.9): D_R[i][j] = 100
        #elif(D_R[i][j] < 0.2 ): D_R[i][j] = -100
        #elif(D_R[i][j] > 5.6 ): D_R[i][j] = 0
        if(  np.abs(ETA(theta[i]*(np.pi/180.0))) < 0.6 and np.abs(ETA(theta[i]*(np.pi/180.0))) > 0.4 ): D_R[i][j] = 0
        elif(np.abs(ETA(theta[i]*(np.pi/180.0))) < 1.6 and np.abs(ETA(theta[i]*(np.pi/180.0))) > 1.4 ): D_R[i][j] = 50
        else: D_R[i][j] = 100

plt.figure(figsize=(10,10))
ax = plt.gca()
ax.set_title("$\Delta$ R")
ax.set_xlabel(r"$\phi(\phi_0$ = 0)")
ax.set_ylabel(r"$\theta(\theta_0$ = 90)")
plt.contourf(X,Y,D_R,60)#,cmap = "RdGy")

plt.show()
```
